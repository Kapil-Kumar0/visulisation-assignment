```python
# 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.
np.random.seed(42)  # For reproducibility
num_samples = 100  # Number of samples
num_features = 5   # Number of features


data = np.random.normal(0, 1, (num_samples, num_features))


data[:, 1] = 0.7 * data[:, 0] + 0.3 * data[:, 1]  # Feature 1 is correlated with Feature 0
data[:, 3] = -0.5 * data[:, 2] + 0.5 * data[:, 3]  # Feature 3 is negatively correlated with Feature 2
# Convert to a pandas DataFrame for easier handling
df = pd.DataFrame(data, columns=[f'Feature {i}' for i in range(num_features)])

# Compute the correlation matrix
corr_matrix = df.corr()


plt.figure(figsize=(8, 6))  # Set the figure size
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, linewidths=0.5)

# Add title
plt.title('Correlation Matrix Heatmap', fontsize=16)
plt.show()
```

```python
# 3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

heatmap_data = df.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')

# Create the heatmap
plt.figure(figsize=(10, 6))  # Set the figure size
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='.0f', linewidths=0.5)

# Add title and labels
plt.title('Variation in Sales Across Months and Days', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Day', fontsize=12)
plt.show()
```

# BOKEH ASSIGNMENT:

```python
# 1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Enable Bokeh to display in a notebook (optional)
output_notebook()

# Generate x and y values
x = np.linspace(0, 10, 500)  # 500 points between 0 and 10
y = np.sin(x)  # Sine of x

# Create a Bokeh figure
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=800, height=400)

# Add a line plot for the sine wave
p.line(x, y, legend_label="y = sin(x)", line_width=2, line_color="blue")

# Add a circle glyph for data points (optional)
p.circle(x, y, size=5, color="red", alpha=0.5)

# Customize the plot
p.title.text_font_size = "16pt"
p.xaxis.axis_label_text_font_size = "12pt"
p.yaxis.axis_label_text_font_size = "12pt"
p.legend.location = "top_right"

# Show the plot
show(p)


```

```python
# 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
#markers based on the 'sizes' and 'colors' columns.




output_notebook()

# Generate random data
np.random.seed(42)
num_points = 50
x = np.random.rand(num_points) * 10  # Random x values between 0 and 10
y = np.random.rand(num_points) * 10  # Random y values between 0 and 10
sizes = np.random.randint(10, 50, num_points)  # Random sizes between 10 and 50
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], num_points)  # Random colors

# Create a DataFrame
df = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

# Create a Bokeh figure
p = figure(title="Scatter Plot with Custom Sizes and Colors", x_axis_label='x', y_axis_label='y', width=800, height=400)

# Add a scatter plot with custom sizes and colors
p.scatter('x', 'y', size='sizes', color='colors', source=df, alpha=0.7, legend_label='Data Points')

# Customize the plot
p.title.text_font_size = "16pt"
p.xaxis.axis_label_text_font_size = "12pt"
p.yaxis.axis_label_text_font_size = "12pt"
p.legend.location = "top_left"

# Show the plot
show(p)
```

```python
# 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

from bokeh.models import ColumnDataSource


output_notebook()

# Dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource
data = {'fruits': fruits, 'counts': counts}
source = ColumnDataSource(data=data)

# Create a Bokeh figure
p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruits', y_axis_label='Counts', width=600, height=400)

# Add a bar chart
p.vbar(x='fruits', top='counts', width=0.5, source=source, color='skyblue', legend_label='Fruit Counts')

# Customize the plot
p.title.text_font_size = "16pt"
p.xaxis.axis_label_text_font_size = "12pt"
p.yaxis.axis_label_text_font_size = "12pt"
p.xgrid.grid_line_color = None  # Remove x-axis grid lines
p.legend.location = "top_right"

# Show the plot
show(p)
```
